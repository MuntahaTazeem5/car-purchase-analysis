***Scatter Plot***

**Problem: **An e-commerce company wants to analyze the relationship between product price and purchases.

**Columns: **price, purchases

**Steps:**

Load sales data.

Select price and purchase columns.

Use sns.scatterplot() to visualize.

Check if high prices reduce purchases.

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('/content/Car.csv')
```

```python
data=pd.DataFrame(data)
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
sns.scatterplot(x='Price_in_thousands',y='sales',data=data)
import matplotlib.pyplot as plt
plt.title("Scatter Plot")
plt.show()
```

```python
correlation = data['Price_in_thousands'].corr(data['sales'])
print(f"The correlation between Price_in_thousands and sales is: {correlation:.2f}")
```
